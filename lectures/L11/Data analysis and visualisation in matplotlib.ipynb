## Ok, but how can I really customise my plots?
To fully customise your plots, you need to use `matplotlib`, the plotting library that
most other Python plotting libraries use behind-the-scenes. Its API is cumbersome to deal
with, but you can make just about any plot you can think of with it!

You may have seen it in action!

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(10*x)*np.exp(-x/2)
plt.plot(x, y)
```

This is the `matplotlib` pyplot API, and it is highly discouraged to use it. Matplotlib was originally
designed to mirror MATLAB's (poorly designed) plotting API. As a consequence, we have a poorly designed
plotting in Python API as well. 

However, much effort has been made in making an easier to use, object oriented API. And it is this
API that I reccomend that you use

```python
fig = plt.figure()
ax = fig.add_subplot(111)  # Equivalent to (1, 1, 1)

ax.plot(x, y)
```

Now, we can customise our figures more!

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111)

ax.plot(x, y)
```

So what happens here?

We use `plt.figure` to create a figure instance (unfortunately, `matplotlib` was created
before PEP-8 was the de-facto style guide for Python, so it does not allways use Pascal case
for class names).

Then, we use `figure.add_subplot` to create an axes-object inside the specified figure. We
then draw our curve inside this axes object.

```python
fig = plt.figure()
ax1 = fig.add_axes((0, 0, 0.8, 1))  # Left, Bottom, Width, Height
ax2 = fig.add_axes((0.8, 0, 0.1, 1))

ax1.plot(x, y)
ax2.plot(y, x)
```

The subplots function is an easy call so you won't have to do arithemtic yourself to figure out the location and limits of the plots.

```python
fig = plt.figure()

all_axes = {
    (2, 3, i+1): fig.add_subplot(2, 3, i+1) for i in range(6)
}

for specification, axes in all_axes.items():
    axes.text(0, 0, f'{specification}', fontsize=18, horizontalalignment='center', verticalalignment='center')
    axes.set_xlim(-1, 1)
    axes.set_ylim(-1, 1)
    axes.set_xticks([])
    axes.set_yticks([])
```

### We can use subplots to create dashboard-like layouts

```python
fig = plt.figure()

all_axes = {
    (1, 2, 1): fig.add_subplot(1, 2, 1),    (2, 3, 3): fig.add_subplot(2, 3, 3),
    (2, 3, 6): fig.add_subplot(2, 3, 6)
}

for specification, axes in all_axes.items():
    axes.text(0, 0, f'{specification}', fontsize=18, horizontalalignment='center', verticalalignment='center')
    axes.set_xlim(-1, 1)
    axes.set_ylim(-1, 1)
    axes.set_xticks([])
    axes.set_yticks([])
```

For each `axes.set_*` method, there is a corresponding `axes.get_*` method that will show you
that attribute

```python
fig = plt.figure()
ax = fig.add_subplot(111)
print(ax.get_xlim())
```

**Imshow can also take heigt x width x 3 RGB images or height x width x 4 RGBA images

```python
xx, yy = np.mgrid[0:1000, 0:1000]
xx, yy = xx*2*np.pi/500 - 0.5*np.pi, yy*2*np.pi/500 - 0.5*np.pi
zz = np.sin(xx) + np.sin(yy)

xx = (xx - xx.min())/(xx.max() - xx.min())
yy = (yy - yy.min())/(yy.max() - yy.min())
zz = (zz - zz.min())/(zz.max() - zz.min())


img = np.stack((xx, yy, zz), axis=-1)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.imshow(img, cmap='magma')
ax.axis('off')  # Remove the axis box and ticks
```

## Interactive plots and GUI programming
We will not learn about GUI programming here, but with IPython widgets, we can learn the basics.

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
slider = widgets.IntSlider(min=0, max=100, value=25, description='test')
display(slider)
```

Now, we register callbacks to this slider. Whenever it changes, we call a specific function.

```python
def printout(event):
    print(event)
slider.observe(printout, 'value')
```

With IPython widgets, we rarely use the observe function to create GUIs, rather, we use the interact function

```python
@widgets.interact(message='message')
def print_message(message):
    print(message)
    print(len(message))
```

Let us use interact for some simple plots

```python
%matplotlib notebook
# Activate interactive matplotlib backend
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(0, 50, 100)

@widgets.interact(power=(0., 10.))
def replot(power):
    ax.clear()
    ax.plot(x, x**power)
```

### When we create more complicated widgets, we often wrap it in classes.

```python
class PowerVisualiser:
    def __init__(self, max_power, xmin, xmax, ymin, ymax): 
        self.power_slider = widgets.IntSlider(min=0, max=max_power, value=int(max_power/2), description='Power')
        self.xmin_slider = widgets.FloatSlider(min=xmin, max=xmax, value=xmin, description='xmin')
        self.xmax_slider = widgets.FloatSlider(min=xmin, max=xmax, value=xmax, description='xmax')
        self.ymin_slider = widgets.FloatSlider(min=ymin, max=ymax, value=ymin, description='ymin')
        self.ymax_slider = widgets.FloatSlider(min=ymin, max=ymax, value=ymax, description='ymax')
        
        self.layout = widgets.VBox(
            [
                widgets.HBox([self.power_slider]),
                widgets.HBox([self.xmin_slider, self.xmax_slider]),
                widgets.HBox([self.ymin_slider, self.ymax_slider])
            ]
        )
        display(self.layout)
        
        self.fig = plt.figure()
        self.ax = self.fig.add_subplot(111)
        
        self.replot()
        for sublayout in self.layout.children:
            for widget in sublayout.children:
                widget.observe(self.replot)

    def replot(self, *args, **kwargs):
        xmin = self.xmin_slider.value
        xmax = self.xmax_slider.value
        ymin = self.ymin_slider.value
        ymax = self.ymax_slider.value
        power = self.power_slider.value
        
        x = np.linspace(xmin, xmax, 1000)
        self.ax.clear()
        self.ax.plot(x, x**power)
        self.ax.set_xlim(xmin, xmax)
        self.ax.set_ylim(ymin, ymax)
```

```python
v = PowerVisualiser(10, -5, 5, -100, 100)
```

## Matplotlib events

We can also capture mouse events within the matplotlib window, see documentation [here](https://matplotlib.org/3.1.1/users/event_handling.html)

```python
# Cheekily modified from the documentation link above

fig, ax = plt.subplots()
ax.plot(np.random.rand(10))

def onclick(event):
    ax.set_title(f'({event.xdata:.2f}, {event.ydata:.2f})')
    
cid = fig.canvas.mpl_connect('button_press_event', onclick)
```

This also works outside of notebooks. Also, note that `print` calls do not work with matplotlib event-handling in a notebook environment.
